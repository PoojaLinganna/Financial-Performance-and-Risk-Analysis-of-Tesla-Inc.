```python
# Install required library
! pip install yfinance pandas openpyxl -- quiet
import yfinance as yf
import pandas as pd
# CHANGE THIS TO YOUR CHOSEN COMPANY
ticker = 'TSLA' # Examples : MSFT , TSLA , KO , JPM , AMZN , NVDA
company_name = yf . Ticker ( ticker ) . info . get ('longName', ticker )
print ( f" Downloading financial statements for { company_name } ({ ticker }) ...\ n")
# Helper function to clean and transpose statements
def prepare_statement ( df , statement_name ) :
  df = df.transpose ()
  df.insert (0 , 'Date', df.index )
  df.reset_index ( drop = True , inplace = True )
  df.fillna (0 , inplace = True )
  print ( f"{ statement_name }␣ downloaded :␣{df. shape [1] -1} ␣ items ␣ ␣{df. shape [0]} ␣ years ")
  return df
# Download all three statements ( annual )
company = yf . Ticker ( ticker )
# 1. Income Statement
income_stmt = company . financials
income_stmt = prepare_statement ( income_stmt , " Income ␣ Statement ")
income_stmt . to_csv (' income_statement . csv ', index = False )
# 2. Balance Sheet
balance_sheet = company . balance_sheet
balance_sheet = prepare_statement ( balance_sheet , " Balance ␣ Sheet ")
balance_sheet . to_csv (' balance_sheet .csv ', index = False )
# 3. Cash Flow Statement
cash_flow = company . cashflow
cash_flow = prepare_statement ( cash_flow , " Cash ␣ Flow ␣ Statement ")
cash_flow . to_csv ('cash_flow .csv ', index = False )
print ("\ nAll ␣3␣ statements ␣ saved !␣ Download ␣ the ␣3␣ CSV ␣ files ␣ from ␣ the ␣ left ␣ panel .")
print (" Files :␣ income_statement .csv␣|␣ balance_sheet . csv ␣|␣ cash_flow . csv ")
```
